```python
img_propaganda = cv2.imread('Assets/propaganda.jpg')
img_draka = cv2.imread('Assets/draka.jpg')

img_propaganda = cv2.cvtColor(img_propaganda, cv2.COLOR_BGR2RGB)
img_draka = cv2.cvtColor(img_draka, cv2.COLOR_BGR2RGB)

gray_propaganda = cv2.cvtColor(img_propaganda, cv2.COLOR_RGB2GRAY)
gray_draka = cv2.cvtColor(img_draka, cv2.COLOR_RGB2GRAY)

gray_propaganda_resized = cv2.resize(gray_propaganda, (gamma_corrected.shape[1], gamma_corrected.shape[0]))
gray_draka_resized = cv2.resize(gray_draka, (gamma_corrected.shape[1], gamma_corrected.shape[0]))
```

Script ini membaca gambar "propaganda" dan "draka", lalu mengubahnya dari format BGR ke RGB agar warnanya sesuai tampilan asli. Setelah itu, kedua gambar dikonversi menjadi grayscale untuk memudahkan pemrosesan berbasis intensitas. Gambar-gambar grayscale tersebut kemudian diubah ukurannya agar memiliki dimensi yang sama dengan gambar "gamma_corrected", sehingga nantinya bisa dilakukan spesifikasi histogram dengan tepat.

```python
def match_histogram(source, reference):
    original_shape = source.shape
    source = source.ravel()
    reference = reference.ravel()

    src_values, src_indices, src_counts = np.unique(source, return_inverse=True, return_counts=True)
    ref_values, ref_counts = np.unique(reference, return_counts=True)

    src_cdf = np.cumsum(src_counts).astype(np.float64) / src_counts.sum()
    ref_cdf = np.cumsum(ref_counts).astype(np.float64) / ref_counts.sum()

    mapped_values = np.interp(src_cdf, ref_cdf, ref_values)
    matched = mapped_values[src_indices].reshape(original_shape)

    return matched.astype('uint8')
```

Fungsi "match_histogram" yang bertujuan untuk mencocokkan distribusi histogram gambar sumber (source) agar menyerupai gambar referensi (reference). Pertama, gambar diratakan menjadi satu dimensi. Fungsi ini menghitung nilai unik dan jumlah kemunculannya, lalu membuat fungsi distribusi kumulatif (CDF) untuk kedua gambar. Dengan menggunakan interpolasi, nilai-nilai dari gambar sumber disesuaikan mengikuti CDF gambar referensi. Hasil akhirnya dikembalikan dalam bentuk array 2D dengan tipe data "uint8".

```python
def adjust_brightness(image, beta):
    image = np.clip(image + beta, 0, 255).astype(np.uint8)
    return image
```

Fungsi "adjust_brightness" untuk mengubah tingkat kecerahan sebuah gambar. Nilai piksel gambar dinaikkan atau diturunkan sesuai parameter "beta", kemudian hasilnya dipotong (clipping) agar tetap berada dalam rentang 0 hingga 255. Gambar yang sudah disesuaikan kecerahannya lalu dikonversi kembali ke tipe data "uint8" agar sesuai dengan format citra standar.

```python
matched_to_propaganda = match_histogram(gamma_corrected, gray_propaganda_resized)
matched_to_draka = match_histogram(gamma_corrected, gray_draka_resized)

plt.figure(figsize=(15,6))

plt.subplot(1,3,1)
plt.imshow(gamma_corrected, cmap='gray')
plt.title('equalized')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(matched_to_propaganda, cmap='gray')
plt.title('Propaganda')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(matched_to_draka, cmap='gray')
plt.title('Draka')
plt.axis('off')

plt.tight_layout()
plt.show()
```
